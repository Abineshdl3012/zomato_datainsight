```python
import mysql.connector
import pandas as pd

class Database:
    def __init__(self, host, user, password, dbname):
        self.host = host
        self.user = user
        self.password = password
        self.dbname = dbname
        self.conn = None
        self.cursor = None

    def connect(self):
        try:
            self.conn = mysql.connector.connect(
                host=self.host,
                user=self.user,
                password=self.password,
                auth_plugin="mysql_native_password"
            )
            self.cursor = self.conn.cursor()
            print("Connected to MySQL database")
        except mysql.connector.Error as e:
            print(f"Error connecting to database: {e}")

    def create_database(self):
        try:
            self.cursor.execute(f"CREATE DATABASE {self.dbname}")
            print(f"Database {self.dbname} created successfully")
        except mysql.connector.Error as e:
            print(f"Error creating database: {e}")

    def create_table(self, table_name, df):
        try:
            self.cursor.execute(f"USE {self.dbname}")
            self.cursor.execute(f"CREATE TABLE {table_name} ({', '.join([f'{col} VARCHAR(1000)' for col in df.columns])})")
            print(f"Table {table_name} created successfully")
        except mysql.connector.Error as e:
            print(f"Error creating table: {e}")

    def insert_data(self, table_name, df):
        try:
            self.cursor.execute(f"USE {self.dbname}")
            for index, row in df.iterrows():
                self.cursor.execute(f"INSERT INTO {table_name} VALUES ({', '.join(['%s'] * len(row))})", tuple(row))
            self.conn.commit()
            print(f"Data inserted into table {table_name} successfully")
        except mysql.connector.Error as e:
            print(f"Error inserting data: {e}")

    def close_connection(self):
        if self.conn:
            self.conn.close()
            print("Database connection closed")

# Create a database
db = Database("localhost", "root", "Abinesh", "zomato")
db.connect()
db.create_database()

# Create tables and insert data
db.create_table("customers", customers_table)
customers_table['customer_id'] = customers_table['customer_id'].apply(lambda x: x[:50])
db.insert_data("customers",  customers_table)

db.create_table("restaurants", restaurants_table)
db.insert_data("restaurants", restaurants_table)

# Create the table
db.create_table("orders", orders_table)
orders_table['order_id'] = orders_table['order_id'].apply(lambda x: x[:50])
db.insert_data("orders", orders_table)

db.create_table("deliveries", deliveries_table)
db.insert_data("deliveries", deliveries_table)
```
